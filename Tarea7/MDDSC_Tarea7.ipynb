# PL - Definiciones breves

## Fluorescencia

La Fluorescencia se define como la emisión de luz por parte de una sutancia u objeto que ha absorbido luz u alguna otra radiación electromagnética. Se trata de un caso particular de fotoluminiscencia, es decir, emisión de luz que no se produce por radiación de cuerpo negro.

En muchos casos, la luz emitida tiene una longitud de onda mayor a la que la produjo y por tanto, menor energía que la radiación absorbida. 

El caso más común de fluorescencia se da cuando la radiación absorbida es ultravioleta (y por lo tanto, no-visible para los humanos) mientras que la luz emitida está dentro del espectro visible. Así la sustancia u objeto fluorescente presenta un color distintivo cuando está bajo la exposición de luz UV.

En el caso de la fluorescencia, la emisión se detiene inmediatamente después de que la fuente de luz absorbida interrumpe su radiación.

## Fosforescencia

La fosforescencia es también un tipo de fotoluminiscencia, aunque diferente a la fluorescencia. A diferencia de la anterior, un material fosforescente no reemite inmediatamente la radiación que absorbe. La demora en la reemisión está asociada con las transiciones a estados de energía prohibidos según las reglas de la mecánica cuántica. Ocurren aquí otros fenómenos como inversiones de spin. Como estas transiciones pueden llegar a ocurrir muy lentamente, por lo que la radiación absorbida puede ser reemitida en una intensidad mucho más baja por varias horas después de la excitación original, incluso si la misma ha cesado.

## Electroluminiscencia

La Electroluminiscencia es un fenómeno optoelectrónico en el cual un determinado material emite luz como respuesta al pasaje de una corriente eléctrica a través del mismo o a ser expuesto a un campo magnétido intenso. La electroluminiscencia es el resultado de la recombinación de electrones y huecos de un semiconductor. Los electrones excitados liberan su energía generando fotones y por lo tanto, luz.

## Electroluminiscencia de Juntura

Cuando en un semiconductor dopado se forma una juntura, los electrones y los huecos se encuentran separados en la zona de la juntura. Así, cuando la EL se da en este caso, se habla de eletroluminiscencia de juntura.

## Catodoluminiscencia

La catodoluminiscencia es un fenómeno optoelectrónico en el cual los electrones que impactan sobre un material como el fósforo, provocan la emisión de fotones que pueden estar en el espectro visible. Un ejemplo típico es el del tubo de rayos catódicos recubierto con fósforo, que al ser excitado por un haz de electrones genera luz en la zona donde estos impactan. La catodoluminiscencia puede pensarse como el fenómeno inverso al efecto fotoeléctrico, en el cual se generan electrones a partir de la estimulación de un material con fotones incidentes sobre el mismo.



# MBE

## Knudsen cell

Una celda de Knudsen es una tipo de celda de efusión utilizada como fuente de evaporación. Se utilizada con elementos de presiones parciales bajas, como Ga, Al, Hg, etc. 

La celda de Knudsen permite un excelente control de la temperatura del material a evaporar, por lo cual resultan muy adecuadas para ser utilizadas en MBEs.

En los sistemas MBE, el haz molecular se genera a partir del calentamiento de un material hasta que se vaporiza, dentro de una celda que posee un muy pequeño orificio de salida. Esta celda se denomina celda de efusión o celda de Knudsen. A continuación reproducimos una figura esquemática.

<img src="img/knudsen.png" width="600">

Una celda de Knudsen típica contiene un crisol (típicamente hecho de Nitruro de Boro, Cuarzo, Tungsteno o Grafito), filamentos calentadores (usualmente de Tantalio), un sistema de refrigeración or agua, aisladores de calor y un shutter que abre o cierra la salida del orifcio de salida.

A medida que el vapor se escapa a través del orificio de efusión (o "nozzle"), sus moléculas (o átomos) van formando un haz bien colimado, dado que las condiciones de ultra alto vacío (UHV) en las que el sistema trabaja, permiten que las moléculas u átomos que escapan por el nozzle formen trayectorias balísticas por largas distancias sin colisiones.

# GSMBE

No todos los sistemas MBE funcionan a partir de celdas de efusión. Los Gas Source MBE (o GSMBE) no utilizan celdas de efusión, sino que los materiales a depositar se encuentran en forma gaseosa. En este caso, el sistema de manejo de las fuentes de gases debe ser especialmente cuidadoso para asegurar la precisión y estabilidad del flujo gaseoso necesario para formar el haz.

# Magnetron Sputtering

Existen básicamente cuatro tipos de magnetron sputtering, dependiendo en la fuente de energía utilizada. Los cuatro tipos son:

* DC Sputtering
* RF Sputtering
* Mid-Frecuency AC Sputtering
* Pulsed DC Sputtering
* HIPMIS

## DC Sputtering

Es la alternativa más sencilla y barata para deposición metálica. La fuente de DC es sencilla de control y barata en comparación con otras. Su control es simple, pudiendo variar así la tasa de depósito y sus características con facilidad.

La configuración básica se muestra a continuación. El blanco o target suele situarse en una cámara de vacío en paralelo al sustrato sobre el cual se desea depositar. El blanco se coloca a un potencial negativo respecto de la referencia de la cámara.

La cámara trabaja a una presión de vacío, luego de lo cual se hace ingresar un gas muy puro e interte (típicamente Argón).

Inicialmente, los átomos neutros del Argón se ionizan como resultado de la colisión de los mismos con la superficie del blanco. El blanco expulsa entonces átomos formando un plasma. 

Los átomos vaporizados generados por el blanco son atraídos hacia el sustrato, que se coloca a un potencial positivo. Los átomos que llegan al sustrato comienzan un proceso de nucleación que va conformando un film delgado. El sistema de Magnetron atrapa los electrones sobre el blanco, evitando que bombardeen el sustrato y mejorando la eficiencia del depósito. 

<img src="img/DC.png" width="400">

Las principal limitación del DC sputtering  es que resulta muy difícil realizar depósitos de materiales aislantes, ya que los mismos se cargan y producen arcos. También puede que el target quede contaminado con una carga tal que no se produzca el proceso de sputtering. Además, si el ánodo se depositara con material aislante, este podría eventualmente desaparecer.

Si se utilizan magnetrones desbalanceados, el sistema permite aplicar una determinada polarización o bias al sustrato, que formará un plasma secundario en su proximidad. Los iones de este plasma se aceleran hacia el sustrato, produciendo en esta zona un plasma mucho más denso en su entorno, generando mejores propiedades en el film depositado. Como contraparte, aumentala carga térmica del sustrato y por lo tanto, suele generar mayore stemperaturas sobre el mismo. Adicionalmente, los magnetrones desbalanceados generan un punto de campo magnético nulo cerca del material, produciendo la "Race track erosion" sobre el blanco, que es una depresión circular en la superficie, desaprovechando buena parte del blanco.

## RF Sputtering

Esta técnica, en lugar de aplicar un potencial continuo, aplica una señal alterna que por lo tanto, invierte la polaridad de ambos electrodes, evitando efecto de carga de los blancos y consecuentes problemas de arcos y contaminación.

Al igual que en el caso de DC, se utiliza un gas inerte en una cámara de vacío, el cual se ioniza. El material es bombardeado por estos átomos ionizados como un "spray" que cubre el sustrato y lo va depositando. También se utilizan imagnes para confinar los electrones sobre el blanco, para que no lleguen al sustrato.

Se utilizan dos electrodos con corriente alterna, lo que implica que la polaridad se alterna sucesivamente, "limpiando" la superficie del blanco en cada inversión, lo que disminuye enormemente el proceso de carga del sustrato, y los consecuentes arcos que podrían dañar al blanco e incluso evitar que se logre el depósito. La frecuencai típica es de 13.56MHz, valor típico de aplicaciones de electrónica de radiofrecuencia.

A diferencia de lo qu ocurre con la fuente de DC, el RF Sputtering tiende a generar un plasma más difuso en toda la cámara más que concentrarlo sobre el blanco. Este plasma puede sostenerse a presiones más bajas, por lo que resulta en menores colisiones de gas ionizado y por lo tanto un depósito más eficiente.

Comparado con las alternativas de DC, el Sputtering de RF genera una menor erosión de tipo Race Track sobre el blanco, y por lo tanto, lo aprovecha mejor, por estar los electrones menos confinados en torno al blanco. Adicionalmente, evita el efecto de desaparción de ánodo.

El RF Sputtering implica costos bastante mayores que el de DC, dado que la electrónica de RF es bastante más cara, así como el control de su potencia. Requiere también el uso de redes de adaptación para evitar problemas de onda reflejada. Las tasas de depósito son consdierablemente más lentas que en el caso de DC y requieren de tensiones mayores. Esto también puede provocar sobrecalentiamiento, y por lo tanto, exige trabajar mejor la refrigeración dle sistema. Por otra parte, el efecto pelicular exige trabajar con conductores y conectores especiales y bastante más complejos.

El RF Sputtering es la técnica más común para el depósito de materiales aislantes.

## Mid-Frecuency AC Sputtering

Esta técnica resulta adecuada para depositar films no conductores. Al operar en frecuencias más bajas que el RF Sputtering (kHz en lugar de MHz) la fuente de alimentación y la electrónica asociada son mucho más simples, logrando un proceso más barato.

Se trabaja en un rango de tensiones de 300V a 1200V, en general entre 25 y 300kW de potencia, a frecuencias que van de los 20 a los 70kHz, siendo 40kHz un valor típico. Una aplicación típica de los MFAC Sputtering es en los sputtering reactivos, que buscan la formación de óxidos durante el proceso de depósito.

Los sistemas más modernos utilizan electrónica para detectar y eliminar arcos, mejorando la estabilidad del sistema e incrementando las tasas de depósito. Además se evita el efecto de la "desaparición del ánodo" por depósito de un aislante, ya que cada electrodo funciona en forma alternativa como ánodo y cátodo cada medio ciclo del proceso.

Los MFAC usan un magnetrón dual para confinar a los electrones en torno al blanco, que puedan ser balanceados o desbalanceados y estar localizados en diferentes configuraciones (lado a lado, inclinados, cara a cara, etc). 

Como resumen, el MFAC Sputtering permite trabajar con potencias menores que que otras fuentes como las de RF, además de ser más barato, estable y permitir mayores tasas de depósito.

## Pulsed DC Sputtering

El Pulsed DC Sputtering permite depositar tanto films conductores como aislantes.

Su uso es muy común cuando se desea realizar Sputterings reactivos, donde se da una reacción química en el plasma entre el material vaporizado y gases ionizados como por ejemplo el oxígeno, para así formar por reacción química moléculas diferentes. Esta técnica permite depositar materiales típicamente complicados como Alumina, Titania y Silica con tasas de depósito que serían inalcanzables con una fuente de DC simple, pues los arcos serían muy graves para este tipo de blancos.

En vez de aplicarse un potencial fijo, se aplican pulsos de tensión. Se trabaja con pulsos de frecuencia de 10 a 350kHz, y ciclos de actividad del 50 al 90%. Durante el tiempo activo o "ton", un pulso negativo de cientos de volts es aplicado al blanco para iniciar el sputtering. El gas en la cámara se ioniza por las colisiones con los átomos y comienza el plasma. El depósito solo ocurre durante este período. Luego del "ton", el voltaje se apaga, o en algunos casos, se aplica incluso una tensión positiva para limpiar el target e impedir cargas. Esta reversiónde voltaje (usualmente de unos 20V) se mantiene entre un diez y un veinte por ciente de la duración del "ton". En este tramo del ciclo, la superficie del blanco se descarga, dejándola lista para el próximo ciclo de plasma.

La principal ventaja respecto a la fuente de DC es que se reduce fuertemente la presencia de arcos. El control no es sencillo y mantener la estabilidad del proceso tampoco lo es, asimismo puede producirse un efecto de carga remanente que podría obligar a rearrancar el proceso luego de algunos minutos.

Los sistemas más modernos incluyen circuitos electrónicos para detectar y apagar arcos, aplicando tensiones de polaridad opuesta.

Aunque el Pulsed DC Sputtering fue originalmente desarrollado por la posibilidad de evitar arcos al depositar materiales dieléctricos, también tiene otros beneficios como la posibilidad de controlar la cantidad de iones en el flujo del plasma, lo que mejora la adhesión y permite controlar otras propiedades del film depositado.

Como resumen, la principal ventaja del Pulsed DC Sputerring es que encontrando los valores óptimos de frecuencia y ciclo de actividad, se pueden lograr grandes tasas de depósito en comparación con el RF Sputtering, sin los problemas de arcos típicos del DC Sputtering.

## HIPIMS

High Power Impulse Magnetron Sputtering o HIPIMS es una de las técnicas más modernas utilizadas para depósitos físicos por Magnetron Sputtering. Utiliza una fuente de tensión pulsada de alta tensión. Se trabaja a voltajes altos, con una ráfaga de energía durante un tiempo breve, enfocada sobre el blanco de material a depositar, de forma tal que se genera un plasma de muy alta densidad que produce un alto grado de ionización del material a depositar.

Trabajando con ráfagas de alto voltaje (de período aproximado de ~100μs, en el orden de los kW⋅cm−2 pero con ciclos de actividad menores al 10%), se logra que una gran parte del material del blanco se ionice en la "nube" de plasma, pero sin sobrecalentar el blanco ni otros componentes del sistema. Durante el tiempo de apagado o "toff", el blanco puede enfriarse, lo que permite una potencia de cátodo baja (1 a 10kW) y por lo tanto, un proceso más controlado y repetible.
La técnica permite generar depósitos mucho más densos, con mejor adherencia a los sustratos y mucho menos rugosos en comparación con las técnicas clásicas de Magnetron Sputtering. 

La combinación de alta tensión y breves períodos de tiempo provocan una mayor velocidad de ionización sobre el material del blanco, resultando así en una mayor adhesión y mejor uniformidad lograda sobre los sustratos a depositar. Esto permite utilizar sustratos de formas más complejas, donde los métodos más comunes no podrían cubrir los picos y valles propios de un sustrato irregular.

En general, el HIPMIS despierta gran interés pues casi cualquier Magnetron Sputtering clásico puede adaptarse a este sistema fácilmente, en general, reemplazando la fuente por una fuente HIPMIS y algunos pocos componentes adicionales.

Cuando el pulso se aplica, se genera un plasma denso con forma de "donut", el cual se espande desde la zona de erosión del blanco hacia el magnetrón. Este plasma en expansión se refleja en las paredes de la cámara, volviendo hacia el centro, generando un segundo pico de intenisdad varios cientos de μs después de que el pulso de excitación haya terminado. La descarga inicial está dominada por los iones del gas, mientras que en la segunda existe una gran componente de electrones. Este confinamiento de electrones secundarios es fundamental para la creación de un plasma denso. Esto genera una ventaja adicional del HIPIMS: la mayor ionización del plasma dada por las colisiones de gases secundarios hace que el plasma resultante sea más frío al alcanzar el sustrato, lo que permite realizar depósitos sobre sustratos sensibles a las altas temperaturas.

Existen dos problemas destacables del HIPIMS:

* Requiere mayor potencia para lograr iguales tasas de depósito que un Magnetron Sputtering tradicional. Estas pueden ser de 30% a 80% mayores.
* Produce arcos sobre el blanco. Las causas de los mismos pueden ser variadas, existiendo materiales mucho más susceptibles a producir arcos que otros.


# PLD

## Ablación por láser pulsado

La duración del pulso láser aplicado puede tener grandes diferencias en el resultado de la ablación.

Duraciones de pulsos en el rango de los femtosegundos y pocos picosegundos se consideran pulsos "ultra cortos". Los pulsos ultra cortos, solo permiten interacción del laser con los electrones. Por otra parte, pulsos de duración más larga, generan interacción con la estructura cristalina o lattice.

Cuando se utilizan pulsos ultra cortos, la conducción del calor está limitada. Como consecuencia de esto, la ablación sobre el material será limitada espacialmente, minimizando el daño mecánico y térmico subrido por el blanco. Por el contrario, pulsos de duración más larga (del orden de los nanosegundos) conlleva un calentamiento más continuo del material del blanco. La energía del pulso se distribuirá entonces por conducción hacia un área que excede el tamaño del haz, lo que podría causar que el blanco irradiado se caliente hasta fundirse.


## Efecto de utilizar diferentes anchos de pulsos

En la ablasión láser, pueden distinguirse básicamente tres etapas:
* Absorción de fotones
* Calentamiento y fotoionización del área del blanco alcanzada por el haz
* Liberación del material del blanco en forma de fragmentos sólidos, vapor, gotas líquidas y/o pluma de plasma

La cantidad de material erosionado y su fase dependen de la energía que el blanco absorbe. Luego de la interacción con pulsos de duración breve y baja intensidad, a causa de la radiación de frenado inversa, la energía del haz láser será absorbida por electrones libres del material, lo que se traducirá en termalización y posterior transferencia de carlor a la estructura cristalina. La interacción térmica entre láser-electrones-lattice hace que variar los tiempos del pulso tenga efectos diferentes en la ablación. 

## Pulsos de femtosegundos

Cuando el pulso láser dura un tiempo de femtosegundos, esto implica que el pulso dura menos que el tiempo de enfriamiento de los electrones. Entonces, el acoplamiento térmico entre los electrones y el lattice puede despreciarse durante el calentamiento.

En este caso, luego de que el pulso cesa, los electrones se enfrían rápidamente por la transferencia de energía hacia la estructura cristalina y transferencia de calor hacia el bulk. 

En la ablación con pulsos de femtosegundos, el proceso de ablación puede considerarse como una transición directa de sólido a vapor, o sólido a plasma. En este caso, el lattice se calienta en un lapso de picosegundos, lo que provoca la creación de vapor y plasma en rápida expansión en condicones de vacío. Durante todos estos procesos, la conducción térmica hacia el blanco puede ser despreciada.

## Pulsos de picosegundos

En este caso, las condiciones cambian, dado que el tiempo del pulso supera el tiempo de enfriamiento de los electrones. La temperatura de los electrones puede pensarse entonces como cuasi-estacionaria.

Para estos pulsos, la temperatura del lattice permanece en un valor mucho menor a la temperatura de los electrones, y el enfriamiento de los electrones se da mayormente por intermabio energético con el lattice. 

En esta caso, la ablación viene acompañada por conducción térmica de los electrones que podrían generar una pequeña zona líquida dentro del blanco. La presencia de la fase líquida dentro del target reduce la precisión de la ablación en este régimen.

## Pulsos de nanosegundos.

En este régimen, la temperatura de los electrones y del lattice puede considerarse igual. En este caso, la energía absorbida calienta la superficie hasta el punto de fusión y luego hasta la vaporización. Como se sabe, los metales necesitan mucha más energía para vaporizarse que para fundirse. Durante la interacción, se produce conducción por el sólido que en este caso, es una pérdida de energía. La profundidad de la penetración del calor depende de la constante de difusión del material. 

Al igual que en el caso anterior, existe una zona de material derretido, pero aquí es todavía más gruesa que en el caso anterior. La evaporación entonces ocurre del metal líquido, lo cual hace muy complicado controlar el proceso de  ablación.


## Comparación de los largos de pulsos

En general, podemos decir que utilizando lasers de nanosegundos la ablación ocurre dada la presión de vapor del material derretido. El material derretido se resolificará después, explicando así la inestabilidad del proceso, en el cual la dinámica entre la fase líquida y el vapor son muy complicadas. Como consecuencia de esto, el área de ablación es poco precisa y uniforme en comparación con la que se produciría con un laser de femto-segundos.

<img src="img/nanofemto.png" width="500">


Adicionalmente, el láser de nanosegundos genera una zona afectado por el calor sobre el blanco (Heat-Affected Zone o HAZ) hasta un orden de magnitud más grande que la generada por un láser de femtosegundo.

Cuando se utilizan láseres de tiempo corto y ultra corto (pico y femtosegundos respectivamente), la duración del pulso es considerablemente menro que el tiempo requerido para la transferencia de energía entre el lattice y los electrones libres del blanco. Como consecuencia de esto, se desarrollan muy altas temperaturas y presiones en una zona muy superficial del blanco, en el orden de pocos micrones. Así, el material irradiado se calienta rápidamente por el láser pulsado y alcanza en forma directa la fase vapor con gran energía cinética, sin pasar por la fase líquida. En otras palabras, la vaporización es casi directa, sin refundir la zona de ablación. Esto produce una HAZ muy pequeña y una ablación muy suave como se observa en la foto anterior. 

En el caso de láseres de femtosegundos, los tiempos son muy cortos incluso en comparación con los procesos de relajación atómica. Después de que la energía se absorbe por los electrones, se producen iones fríos, llevando esto a la termalización al final del pulso láser. Además, la intensidad es grande y suficiente para generar procesos de absorción no lineales. A estas intensidades, ocurre ionización de multifotones que produzcan una ionización en avalancha, en el que la cantidad de electrones libres crece exponencialmente. 

La cantidad de material removida durante la ablación depende de la cantidad de energía bsorbida por el bulk del blanco de material. La disipación de la ernergía absorbida usualmente ocurre luego de que el pulso láser cesa. Mayormente, dos mecanismos permiten explicar la remoción del material por efecto del láser:
* Vaporización térmica, donde la temperatura de una zona localizada se incrementa por encima de la temperatura de vaporización dadas las colisiones entre electrones y fonones
* Explosión de Coulomb, donde el material bulk libera los electrones excitados y produce un campo eléctrico intenso que remueve los iones del área de impacto.


Como se muestra a continuación, los pulsos lentos tienen una HAZ mayor y una zona de material derretido.

Otra diferencia importante es que para la ablación con laser de nanosegundos, el plasma se produce durante el pulso laser, y en contraste con esto, para duraciones de femtosegundos el plasma se produce luego de que el pulso finaliza. En el primer caso entonces, el pulso láser recalienta el propio plasma. Esto conlleva mayor persistencia del plasma.
