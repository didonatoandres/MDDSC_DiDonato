# Software para XRR

## Remagx

https://remagx.org/wiki/doku.php

Features

* Three different algorithms for reflectivity simulation for optical isotropic and anisotropic media (Parratt/Matrix(Zak)/Full Matrix)
* Graphical User Interface
* Sample definition either as compound-profile or elemental-profile
* Script language for maximum flexibility
* Parallelized algorithms for best performance
* Adaptive slicing/layer segmentation module
* Internal database of refractive index for off-resonant scattering
* Processes optical-constant database files for maximum flexibility
* Variable x-ray polarizations and polarization filters
* Magnetism on a per-layer base or by introducing artificial magnetic moments
* Simulation of Tranmission
* Four different fit routines (Simplex, Genetic Algorithm, Levenberg-Marquardt, Simulated Annealing)
* Map calculation module
* Four different numerical precision modes
* Data import modul 

## Motofit

https://sourceforge.net/p/motofit/wiki/Home/

Motofit co-refines Neutron and X-ray reflectometry data, using the Abeles matrix / Parratt recursion and least squares fitting (Genetic algorithm or Levenberg Marquardt). It works in the IGOR Pro environment (TM Wavemetrics).

Features

* Entering parameters via the reflectivity panel
* Estimating layer thickness from fourier transformation
* Genetic Optimisation of reflectometry data
* Chi2 variation as a function of 2 different parameters
* Fitting reflectivity from a Ti/Ni multilayer with 25 repeats
* Co-refinement of multiple contrast datasets

## IMD

http://www.rxollc.com/idl/

A computer program called IMD is described. IMD is used for modeling the optical properties (reflectance, transmittance, electric-field intensities, etc.) of multilayer films, i.e., films consistingof any number of layers of any thickness.

Features

IMD includes a full graphical user interface and affordsmodeling with up to eight simultaneous independent variables, as well as parameter estimation~including confidence interval generation!using nonlinear, least-squares curve fitting touser-supplied experimental optical data. The computation methods and user interface aredescribed, and numerous examples are presented that illustrate some of IMD’s unique modeling,fitting, and visualization capabilities.


## GenX

http://genx.sourceforge.net/

GenX is a tool to refine x-ray, neutron reflectivity data as well as surface x-ray diffraction data written in Python. The refinement is conducted with an optimization algorithm called differential evolution which is very robust. The model in GenX is defined through a scripting. However, for reflectivity measurements a graphical plugin is available to make it more user friendly.


# Mecanismos de trasnporte en juntura metal-semiconductor

Esencialmente existen cuatro mecanismos de transporte:

* Corrientes de arrastre o deriva (Drift). Es la que se da por el movimiento de portadores de carga dada la aplicación de un campo eléctrico, en general dada por la aplicación de una fuerza electromotriz a una cierta distancia.
* Corrientes de difusión (Diffusion current) Esta corriente se da en los semiconductores y es casuada por la difusión de portadores de carga (electrones o huecos). Se produce dada la no-uniformidad de concentración de las partículas cargadas. Puede oponerse o reforzar a la corriente de deriva y se describe a partir de la ecuación de difusión.
* Emisión termoiónica (Thermoionic emission). La emisión termoiónica es el flujo de portadores inducido a través de una barrera de potencial, cuando la energía térmica aplicada sobre los mismos suepra la función trabajo del material.
* Corrientes de tunelamiento (Tunneling current). El tunelamiento es un efecto de la mecánica cuántica. La corriente de tunelamiento ocurre cuando los electrones pasan a través de una barrera de potencial que no deberían poder atravesar de acuerdo a la física clásica. En térmicos clásicos, si el electrón no tiene la energía suficiente nunca podría pasar a traveś de dicha barrera. Sin embargo, la mecánica cuántica expresa que, las propiedades ondulatorias de los electrones hacen que no dejen de existir en forma abrupta ante una barrera de potencial, sino que la probabilidad de encontrarlos disminuya rápidamente. Si la barrera es suficientemente fina, la probabilidad de existencia podría continuar del otro lado de la misma. Dado que existe una probabilidad pequeña pero no nula, de existir una cantidad suficiente de electrones, efectivamente algunos de ellos pasarán al otro lado.

## Mecanismos que prevalencen en una juntura metal-semiconductor en equilibrio, directa e inversa.

<img src="img/schottky.png" width="800">

En la figura a se muestra una juntura Schottky tipo n en condicones de equilibrio. El quierbe en las bandas se da por la diferencia entre las funciones trabajo de los dos materiales. Los electrones deben ser activados térmicamente para pasar la barrera. Este fenómeno es el de emisión termoiónica. No hay flujo de corriente neto, dado que la corriente desde el metal hacia el semiconductor y desde el semiconductor hacia el metal se cancelan. 

En la figura b, se muestra la nueva condición de las bandas cuando la juntura se polariza en directa. Como se pude ver, la barrera en el lado del semiconductor (que es la que los electrones deben atravesar) se hace mucho más pequeña que en condiciones de equilibrio. La corriente ahora desde el lado del semiconductor es mucho mayor que aquella desde el lado del metal, apareciendo un flujo neto de portadores, y por lo tanto, fluye corriente mayormente por arrastre.

En la figura c), se muestra una juntura Schottky fuertemente polarizada en inversa. Bajo estas condiciones, se dan dos efectos. Uno de ellos es que hay posibilidad de tunelamiento de electrones a través de la barrera de potencial, los cuales se recombinan con electrones en la zona de depleción, emitiendo fotones. También en este modo es apreciable (y en muchos casos dominante) la emisión termoiónica, solo que en este caso la corriente desde el semiconductor hacia el metal es nula, por lo que también existe flujo neto, en sentido contrario a la de la polarización en directa.

# Tipos de Curva IV

## Con diferentes excitaciones
La curva IV se puede realizar aplicando corriente o tensión. Si se aplica una corriente determinada y se mide la tensión, debe cuidarse el límite máximo de tensión, ya que en caso de observarse una impedancia muy alta puede llegar a desarrollarse una tensión muy elevada con consecuentes fenómenos de ruptura dieléctrica o arcos entre electrodos.

## A un puerto o dos puertos
Comunmente llamada "de dos puntas o de cuatro puntas". La medición a cuatro puntas permite disminuir el error de método por caída de tensión en los conductores. A la vez permite graficar curvas IV que representen transresistencias o transadmitancias además de impedancias de excitación.

## Con diferentes configuraciones de conexión de amperímetro/voltímetro
Comunmente se las llama configuraciones de "tensión bien medida" y "corriente bien medida". Permiten disminuir el error de método de la medición directa cuando no se cuenta con configuración de cuatro puntas.

## IV AC 
Se realiza el barrido aumentando la amplitud de una señal alterna senoidal. En rigor de verdad, la gráfica corresponde al módulo de la admitancia a la frecuencia del generador de AC. Se deben tomar especiales recaudos dadas las limitaciones del método en caso de aplicarse a dispositivos no lineales.

## Con diferentes configuraciones temporales
Permiten configurar el tiempo de inicio de medición respecto al inicio del pulso de separación, el tiempo de integración, el tiempo de relajación hasta la próxima medición, etc. Adicionalmente se pueden realizar diferentes tipos de filtrado sobre las muestras capturadas, tales como filtros de media móvil o de mediana.

## En diferentes cuadrantes
Permiten medir diferentes tipos de dispositivos. Los pasivos solo trabajan en cuadrantes 1 y 3, es decir que se comportan siempre consumiendo energía, cualquiera sea la polaridad empleada, mientras que dispositivos activos o que devuelvan energía a la fuente por algún tipo de fenómeno pueden funcionar en los cuatro cuadrantes.

## En forma histerética
Cuando la curva IV se realiza recorriendo valores y volviendo por sobre ellos, pueden observarse fenómenos de histéresis, atribuibles a diferentes efectos físicos. Ejemplos de comportamientos histeréticos se observan cuando un dispositivo retiene cargas. Es también de suma importancia en dispositivos con efecto "memoria", como son por ejemplo los memristores.

## Pulsada
En este caso, en vez de aplicar una tensión de polarización constante se aplica un breve pulso de tensión de muy corta duración y se mide la corriente también en forma pulsada. Aunque pueden existir diferentes motivaciones para realizar una característica IV en forma pulsada, en forma general se suele utilizar en dispositivos que poseen un consumo de potencia apreciable, dado que en estos casos una curva IV estática traería aparejados fenónemos de calentamiento del dispositivo por la propia disipación, los cuales distorsionarían la medición. Con una medición IV pulsada evitamos el calentamiento sin necesidad de trabajar en el sistema de disipación térmica.



# Curvas CV

## LCR Meter


En esencia, un LCR meter mide a partir de un puente de Wheatstone, con uno de sus brazos conectados a un circuito RC paralelo. El componente que se desea medir se conecta en el otro bazo del puente.

A continuación se reproduce un diagrama en bloques.

<img src="img/LCR.jpg" width="600">


Cuando se desea medir un resistor (o la componente resistiva de un dipolo), el puente compara contra su resistencia de referencia. Cuando se busca medir capacitores o inductores, la comparación se realiza con el capacitor de referencia conectado en el brazo RC.

El puente puede medir en DC o en AC, modificándose la excitación del mismo. También se puede medir en DC aplicando una tensión de polarización. El oscilador que se ve en el diagrama en bloques aporta la excitación de AC. Valores típicos son de f=1kHz y f=0.159Hz (ya que implica un valor de pulsación angular unitaria), aunque los LCR meter modernos permiten una variación fina de la frecuencia que típicamente va de unos pocos Hz a algunos MHz.

Para realizar la medición, en forma automatizada el medidor modifica los valores de R y C hasta balancear el puente. La tensión de salida del puente es colectada por un circuito seguidor y luego llega a un detector.

### Limitaciones de los LCR Meters

* Alinealidad de los dipolos medidos. En forma general, estos instrumentos asumen que los dipolos a medir se comportan en forma lineal para el valor de AC utilizado. Un componente alineal produciría una señal de AC de fuerte contenido armónico el cual redundaría en una estimación incorrecta del valor del componente.

* Frecuencia. El método de medición no permite trabajar a frecuencias demasiado grandes, donde la formación de ondas estacionarias por desadaptación distorsionaría las tensiones medidas. Cuando se desea trabajar a frecuencias que superan las decenas de MHz, lo recomendado es utilizar medidores vectoriales que trabajan con cables y conectores adaptados a una impedancia de referencia, y justamente hacen uso de ondas incidentes y reflejadas para la determinación de la impedancia.

* Rangos de valores. Valores muy pequeños o muy grandes de resistencia, capacidad e inductancia no pueden ser medidos pues quedarían apantallados por comportamientos parásitos del sistema, a saber:
    * Resistencia de los conductores
    * Fugas en la aislación
    * Capacidad parásita de conectores

### Quasi-Static CV o QSCV

Este método no utiliza corriente alterna, sino que solo utiliza DC. En rigor de verdad, la variación es muy lenta, y por eso se las denomina cuasi estátitcas. En este tipo de medición se realiza una secuencia de pasos donde se varía la tensión de continua aplicada (como si fuese una señal escalonada) y se mide la corriente resultante, asociada a la carga. Existen dos métodos clásicos como el método de la carga y el método de la rampa. 

Una aplicación típica es la medición de trampas de interfaz. El hecho de trabajar a una frecuencia casi nula, permite obtener diferencias de las curvas en la zona de depleción, pudiéndose inferir a partir de la diferencia entre la curva CV y una QSCV la densidad de trampas en la interfaz (DIT).

El método de la rampa funciona cargando al dispositivo hasta una tensión de DC determinada, utilizando una SMU como fuente de corriente. Una vez que el dispositivo está cargado, una corriente de la polaridad opuesta se hace circular a través de él para descargarlo, mientras que la SMU mide la tensión en función del tiempo. Una segunda SMU mide la corriente de descarga. De las medición de tensión y corriente en función del tiempo, se deriva el valor de capacidad, dado que el mismo tiene una incidencia directa en la pendiente de la rampa de carga y en la constante de tiempo de la exponencial de descarga.

### Very low frequency CV


Es una medición útil para dipositivos como MOScaps, TFTs y MEMS. Las mediciones a muy baja frecuencia que permite este método permite caracterizar los fenómenos de trapping y detrapping, que son de naturaleza muy lenta. Cuando la medición de QSCV no alcanza, se aplica este método que trabaja en frecuencias de 10mHz a 10Hz.

El método de VLF-CV permite medir capacidades muy pequeñas con gran precisión a frecuencias de trabajo muy bajas. Permite medir impedancias muy grandes incluso mayores a 1E15$\Omega$. Las mediciones de curva IV típica alcanzan valores entre 1E6$\Omega$ y 1E9$\Omega$, dependiendo de la calidad del instrumento utilizado. Además, la técnica VLF-CV permite reducir el ruido que se tendría en una medición QSCV, y por ello permite rangos de impedancias tan grandes.

El esquema de medición se muestra a continuación.

<img src="img/vlfcv.png" width="400">

La SMU1 genera una señal de polarización de DC y una señal de AC superpuesta de muy baja frecuencia. Esta también mide la tensión del dispositivo bajo prueba (o DUT por sus siglas en inglés, device under test). La SMU2 mide la corriente de alterna mientras trabaja con una tensión de polarización nula.

Esencialmente, con la tensión forzada, se realizan diversas mediciones de tensión y corriente durante varios ciclos. El módulo y la fase de la impedancia se obtienen a partir de la realización de una Transformada Discreta de Fourier (DFT, mal llamada FFT por su algoritmo de implementación) como una relación de variable compleja entre tensión y corriente senoidal. Esta información de banda angosta se recolecta en un amplio rango de frecuencias (10mHz a 10Hz aproximadamente) para crear un modelo circuital de variable compleja del elemento bajo prueba. El resultado final incluye valore sde impedancia (Z), ángulo de fase ($\theta$), capacidad (C), conductancia (G), resistencia (R), reactancia (X) y factor de disipación (D, inversa numérica del factor de calidad Q).

Como el rango de frecuencias es liimitado, la capadicad del DUT debe estar en el rango de 1pF a 10nF. Adicionalmente, por utilizar la DFT, el método tiene apreciable ruido de fase si el factor de disipación del dipolo es muy bajo. Este esquema se asemeja mucho más a la medición de un impedancímetro (cálculo separado de parte real e imaginaria y posterior composición de la impedancia) que a la realizada por los tradicionales puentes de los LCR meter.
