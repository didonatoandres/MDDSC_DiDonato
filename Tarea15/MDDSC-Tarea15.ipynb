# Sensores basados en MOS

## CMOS CCD

Es un tipo de sensor óptico basado en el efecto fotoeléctrico. En general, se compone de varios fotositos, uno por píxel, los cuales inducen una determinada corriente por efecto fotoeléctrico en función de la intensidad luminosa recibida. La gran ventaja de la tecnología CMOS es que permite integrar gran parte de la electrónica necesaria para generar la imagen en el mismo chip.

## Sensores de gas basados en MOS

Un caso bastante común son los sensores basados en óxidos de estaño (Tin oxide).

<img src="img/gas.png" width="400">

Con aire limpio, electrones donores en el óxido son atraídos hacia el oxígeno, el cual se adsorbe en la superficie del material sensor, impidiendo el flujo de corriente eléctrica.

En presencia de gases reductores, la densidad superficial de oxígeno adsorbido decrece, a medida que reacciona con los gases de reducción. Los electrones entonces se liberan en el óxido, permitiendo circulación de corriente.

## Sensores de ionización MOS

Otra aplicación es la detección de radiación ionizante utilizando una estructura MOS.

Cuando la radiación ionizante pasa a través del dispositivo MOS, pares electrón hueco se crean en el óxido. Dado el campo eléctrio intrínseco o bien, por polarización externa, los electrones se desplazan hasta llegar a encontrar huecos, viajando a través del semiconductor, donde son atrapados en la zona cercana a la interfaz $SiO-SiO_2$. Esto redunda en una variación semipermanente en la tensión de umbral del dispositivo. Las trampas solo se liberan en tiempos muy largos o en condiciones de grandes temperaturas. La cantidad de pares electrón hueco generados es proporcional al espesor del óxido, dado a que se tiene un mayor volumen.

# Sensores

## Sensores resistivos

### LDR (Light Dependent Resistor)

También llamado fotorresistor o fotorresistencia es un componente electrónico cuya resistencia disminuye con el aumento de intensidad de luz incidente.

<img src="img/ldr.jpg" width="300">

El valor de resistencia eléctrica de un LDR es bajo cuando hay luz incidiendo en él (puede descender hasta 50 ohms) y muy alto cuando está a oscuras (varios megaohms). Su funcionamiento se basa en el efecto fotoeléctrico. 

Un fotorresistor está hecho de un semiconductor de alta resistencia como el sulfuro de cadmio, CdS. Si la luz que incide en el dispositivo es de alta frecuencia, los fotones son absorbidos por las elasticidades del semiconductor dando a los electrones la suficiente energía para saltar la banda de conducción. El electrón libre que resulta, y su hueco asociado, conducen la electricidad, de tal modo que disminuye la resistencia. Los valores típicos varían entre 1 MΩ, o más, en la oscuridad y 100 Ω con luz brillante.

Las células de sulfuro del cadmio se basan en la capacidad del cadmio de variar su resistencia según la cantidad de luz que incide en la célula. Cuanta más luz incide, más baja es la resistencia. Las células son también capaces de reaccionar a una amplia gama de frecuencias, incluyendo infrarrojo (IR), luz visible, y ultravioleta (UV).

### Thermistor

Un termistor es un sensor de temperatura por variación de resistencia. Su funcionamiento se basa en la variación de la resistividad que presenta un semiconductor con la temperatura. El término termistor proviene de Thermally Sensitive Resistor.

Existen dos tipos fundamentales de termistores:

* NTC (Negative Temperature Coefficient), donde la resistencia disminuye a medida que aumenta la temperatura.
* PTC (Positive Temperature Coefficient), los cuales incrementan su resistencia a medida que aumenta la temperatura

Los termistores de coeficiente de temperatura negativo (NTC) se usan habitualmente como sensores de temperatura o como limitadores de corriente de arranque. Los termistores de coeficiente de temperatura positivo (PTC) se instalan habitualmente para proteger contra condiciones de sobrecorriente, por ejemplo como fusibles reajustables. El funcionamiento se basa en la variación de la resistencia del semiconductor debido al cambio de la temperatura ambiente, creando una variación en la concentración de portadores. Para los termistores NTC, al aumentar la temperatura, aumentará también la concentración de portadores, por lo que la resistencia será menor, de ahí que el coeficiente sea negativo. Para los termistores PTC, en el caso de un semiconductor con un dopado muy intenso, éste adquirirá propiedades metálicas, tomando un coeficiente positivo en un margen de temperatura limitado. Usualmente, los termistores se fabrican a partir de óxidos semiconductores, tales como el óxido férrico, el óxido de níquel, o el óxido de cobalto.

### Strain Gage o extensómetro

Una extensómetro es un sensor que mide la deformación, presión, carga, par, posición, etcétera, y se basa en el efecto piezorresistivo, que es la propiedad que tienen ciertos materiales de cambiar el valor nominal de su resistencia cuando se les somete a ciertos esfuerzos y se deforman en dirección de los ejes mecánicos. Un esfuerzo que deforma la galga producirá una variación en su resistencia eléctrica. Esta variación se produce por el cambio de longitud, el cambio originado en la sección o el cambio generado en la resistividad. 

En su forma más común, consiste en un estampado de una lámina metálica fijada a una base flexible y aislante. La galga se adhiere al objeto cuya deformación se quiere estudiar mediante un adhesivo, como el cianoacrilato. Según se deforma el objeto, también lo hace la lámina, provocando así una variación en su resistencia eléctrica.

## Sensores capacitivos

### Sensores de desplazamiento y contacto

Se trata de sensores que pueden medir sin contacto directo la posición con gran precisión, de cualquier objeto que sea conductor. También pueden utilizarse para medir el espesor o densidad de materiales aislantes.

En  su variante más básica, una de las caras aislantes se depositan con material conductor. Una pequeña tensión aplicada genera un campo eléctrico uniforme. Cuando un conductor toca la zona no depositada con conductor, se forma un capacitor dinámico. Dada la resistencia superficial de la superficie, la capacidad varía según en qué zona se haya producido el contacto. Así puede determinarse la localización del contacto utilizando un controlador.

<img src="img/touch.png" width="200">

### Sensado de fuerza

Un capacitor de sensado de fuerza es un dispositivo cuya capacidad cambia cuando se le aplica fuerza, presión o stress mecánico. Resultan más sensibles y con respuestas maś repetibles que las que se logran empleando resistores.

En su configuración más típica, se trata de un capacitor de placas paralelas al que, tras aplicarle stress, se deforma una de sus caras. Para deflexiones pequeñas, la variación de capacidad es lineal repecto de la fuerza aplicada.

### Sensado de humedad (Higrómetro)

No son higrómetros de gran precisión, pero los sensores capacitivos de humedad poseen bajo costo, pequeño tamaño y buena resistencia mecánica. Esencialmente explotan el efecto que la humedad tiene sobre el valor de la constante dieléctrica, de un polímero o de un óxido metálico. Con calibración, pueden alcanzar una precisión del 2%.

Resultan muy robustos respecto a efectos como la condensación y el calentamiento.


### Acelerómetros capacitivos

Se basan en tecnología MEMS (Micro Electro Mechanical Systems).

<img src="img/accel.png" width="400">

En el caso más simple se trata de una estructura interdigitada, con un electrodo fijo y uno móvil. La aceleración genera un movimiento oscilatorio que modifica la capacidad medida entre electrodos.

## Sensores Schottky

### Sensado de gas

Utilizan el principio de "Schottky barrier" y se basane n óxidos de estaño y titanio, contactados por metales.

Referencia: Schottky-barrier and conductivity gas sensors based upon Pd/SnO2 and Pt/TiO2

### Sensado de temperatura

Se puden utilizar los diodos Schottky para la medición de temperatura en rangos bajos. Esencialmente se debe regular la corriente termoiónica del diodo para que su variación con la temperatura sea lineal. Se logran sensibilidades interesantes en los valores cercanos a Temperatura ambiente.
 
Referencia: Dutta, Sutanu & Sinha, Sayantan & Panda, Arijit. (2015). Application of a Schottky Diode as a Temperature Sensor. 2350-0352.

###  Fotodiodos

Los fotodiodos tipo Schottky son muy comunes y muy efectivos. Se trata de una estructura muy simple, en la que no se necesita una juntura p-n, por lo que no hay pérdidas ópticas en la capa "p".

La interfaz metal-semiconductor puede lograrse muy abrupta, por lo que la región activa de detección es muy fina. Esto contribuye a que el tiempo de tránsito de los portadores es muy bajo, y por lo tanto, los dispositivos son de muy alta velocidad de respuesta en comparación con otros fotodiodos.

<img src="img/photo.png" width="400">

Esencialmente la energía de los fotones se utiliza para permitir que electrones salten de la banda de valencia a la de conducción, generándose así pares electrón-hueco, siempre manteniendo la polarización inversa del diodo Schottky.

# Interface traps

## GaAs

### Causas

En el caso de los semiconductores III-V y en especial en GaAs, hay dos causas principales de defectos.

* Lattice mismatching o strain. 

<img src="img/strain.png" width="600">

Cuando la constante de Lattice de un sustrato cristalino difiere de la constante de Lattice del material epitaxial crecido, hablamos de un mismatch. A menos que el depósito pueda permanecer "relajado", esto conllevará a un "strain" o tensión que se genera sobre el depósito para adecuarse al sustrato. Este fenómeno existe tanto en silicio como en materiales III-V como el GaAs, pero es mucho más significativo en estos últimos.

* Química de los enlaces

Los átomos de Si o Ge tienen cuatro electrones de valencia, es decir que sus orbitales sp<sup>3</sup> híbridos forman lazos con un electrón. En cambio, en el caso del GaAs, por ejemplo, para el Galio (con tres electrones de valencia) los sp<sup>3</sup> híbridos tendrá 3/4 de electrón, y el Arsénico (con cinco electrones de valencia) tendrá 5/4 de electrón. Cuando se utiliza bulk, esta hibridación funciona adecuadamente ya que cada interfaz híbrida Ga-As contendrá en total 8/4, es decir, dos electrones. En cambio, cuando se forma algún tipo de óxido, estos orbitales híbridos no encuentran un enlace formado por una cantidad entera de electrones, apareciendo así estados no deseados. Los aislantes utilizados para electrónica de GaAs son un tema no-trivial.

En muchos semiconductores III-V, los defectos se ubican en la zona media del bandgap.

Referencia:
C. W. Wilmsen, Ed., Physics and Chemistry of III–V Compound Semicon-
ductor Interfaces. New York: Plenum, 1985.

### Densidad típica

Los valores típicos de densidad de Dit son de 10<sup>11</sup> a 10<sup>13</sup> cm<sup>-2</sup> eV<sup>-1</sup>.

Referencia: IEEE TRANSACTIONS ON ELECTRON DEVICES, VOL. 57, NO. 11, NOVEMBER 2010, Suitability Study of Oxide/Gallium Arsenide Interfaces for MOSFET Applications, Matthias Passlack, Ravi Droopad.

## GaN

### Causas

En el caso del GaN, los "trap states" son básicamente atribuidos a tres causas:

* Interfaces demasiado abruptas
* Enlaces atómicos incompletos
* Presencia de impurezas

Referencia:
A. M. Sonnet, C. L. Hinkle, H. Dawei, and G. Bersuker, IEEE Trans. Electron Devices 57, 2599 (2010).
